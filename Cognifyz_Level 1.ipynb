__Level - 1 Task - 1
  Task: Data Exploration and Preprocessing__

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
re = pd.read_csv("cognify.csv")
re.head()
```

```python
re.shape
```

```python
# Exploring the dataset

print("Number of rows and columns:")
print(re.shape)
```

```python
# Statistical Summary

print("\nStatistical Summary:")
print(re.describe())
```

```python
# Checking for missing values

print("\nMissing values in each column:")
print(re.isnull().sum())
```

```python
#Handling missing values

re['Cuisines'] = re['Cuisines'].fillna(re['Cuisines'].mode()[0])
```

```python
#Checking again for missing values

print("\nMissing values in each column:")
print(re.isnull().sum())
```

```python
# Unique Values

print("\nNumber of unique values in each column:")
print(re.nunique())
```

```python
# Analyzing the distribution of the target variable

plt.figure(figsize=(8, 6))
sns.countplot(re['Aggregate rating'])
plt.title('Distribution of Aggregate Ratings')
plt.xlabel('Ratings')
plt.ylabel('Count')
plt.show()
```

```python
# Checking for class imbalances

class_counts = re['Aggregate rating'].value_counts()
print("\nClass Imbalances:")
print(class_counts)
```

__Level - 1 Task - 2
Task: Descriptive Analysis__

```python
#Getting summary statistics for numerical columns

numerical_stats = re.describe()
print(numerical_stats)
```

```python
# For Country Code

country_counts = re['Country Code'].value_counts()
print("Country Code Distribution:")
print(country_counts)
```

```python
# For City

city_counts = re['City'].value_counts()
print("\nCity Distribution:")
print(city_counts)
```

```python
#For Cuisines

cuisine_counts = re['Cuisines'].str.split(', ', expand=True).stack().value_counts()
print("\nCuisine Distribution:")
print(cuisine_counts)
```

```python
# Top cuisines

top_cuisines = cuisine_counts.head(10)
print("\nTop Cuisines:")
print(top_cuisines)
```

```python
# Top cities with highest number of restaurants

top_cities = city_counts.head(10)
print("\nTop Cities with Highest Number of Restaurants:")
print(top_cities)
```

__Level - 1 Task - 3
Task: Geospatial Analysis

```python
# Plotting restaurant locations

plt.figure(figsize=(10, 6))
plt.scatter(re['Longitude'], re['Latitude'], alpha=0.5)
plt.title('Restaurant Locations')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.grid(True)
plt.show()
```

```python
# Analyzing distribution of restaurants by city

city_counts = re['City'].value_counts()
plt.figure(figsize=(15, 9))
city_counts.plot(kind='bar')
plt.title('Number of Restaurants by City')
plt.xlabel('City')
plt.ylabel('Number of Restaurants')
plt.xticks(rotation=45)
plt.show()
```

```python
# Analyzing distribution of restaurants by country

country_counts = re['Country Code'].value_counts()
plt.figure(figsize=(10, 6))
country_counts.plot(kind='pie', autopct='%1.1f%%')
plt.title('Distribution of Restaurants by Country')
plt.ylabel('')
plt.show()
```

```python
# Correlation between restaurant location and rating 

plt.figure(figsize=(10, 6))
plt.scatter(re['Longitude'], re['Latitude'], c=re['Aggregate rating'], cmap='viridis', alpha=0.5)
plt.colorbar(label='Aggregate Rating')
plt.title('Restaurant Locations with Ratings')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.grid(True)
plt.show()
```
